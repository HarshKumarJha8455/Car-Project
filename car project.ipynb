```python
# TODO: Import Pandas, numpy, matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# TODO: Read the csv file to dataframe
data = pd.read_csv('mtcars.csv')
data
```

This is the information about the columns of the csv file
* mpg	- Miles/(US) gallon
* cyl	- Number of cylinders
* disp - Displacement (cu.in.)
* hp - Gross horsepower
* drat - Rear axle ratio
* wt - Weight (1000 lbs)
* qsec - 1/4 mile time
* vs - Engine (0 = V-shaped, 1 = straight)
* am - Transmission (0 = automatic, 1 = manual)
* gear - Number of forward gears

Let's see the realtionship between engine displacement and milage of the cars

```python
# Plot a scatter plot of disp vs mpg
x=data[['disp']].values
y=data[['mpg']].values
plt.scatter(x,y)
plt.xlabel('Displacememt')
plt.ylabel('Milage')
plt.show()
```

You can see that there is a co-relation between milage and displacement, i.e, generally milage decreases when displacement increases.

Training a Linear regression model to the above data between the displacement and milage.
Taking X = disp
and, Y = mpg

```python
# Selecting the specific columns for X and Y from dataframe
X = data[['disp']].values
Y = data[['mpg']].values
```

```python
# Converting X and Y to Numpy array
X = np.array(data.disp)
y = np.array(data.mpg)
X,y
```

```python
# importing sklearn & build a regression Model on disp and mpg
from sklearn.linear_model import LinearRegression
```

```python
X = X.reshape(-1,1)
```

```python
reg = LinearRegression().fit(X, y)
```

```python
# R2 Score
reg.score(X, y)
```

```python
# Slope
reg.coef_
```

```python
# Intercept
reg.intercept_
```

```python
# function for plotting data points
def plot_points(X, y, xlabel, ylabel):
    """Plot points given X and Y co-ordinates and labels them"""
    plt.plot(X, y, 'o')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
# plots line of regression
def draw_line(X, m, c):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = X
    y_vals = m * X + c
    plt.plot(x_vals, y_vals, '--')
```

```python
# plots points and lines
def draw_points_and_lines(X, y, xlabel, ylabel, m, c):
    """Draws lines and points"""
    plot_points(X, y, xlabel, ylabel)
    draw_line(X, m, c)
```

```python
# draw line of regression without traing model
draw_points_and_lines(X.flatten(), y, "Displacement", "Milega", reg.coef_, reg.intercept_)
```

```python
#Making a prediction of Y(mpg) at X=200.0
reg.predict([[200.0]])
```
